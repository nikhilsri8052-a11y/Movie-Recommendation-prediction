```python
import numpy as np
import pandas as pd
import difflib # for Fuzzy name matching. (string-matching technique that identifies similar and non-identical names in datasets)
from sklearn.feature_extraction.text import TfidfVectorizer # concerting text into numeric vectors.
from sklearn.metrics.pairwise import cosine_similarity # computing similarity
```

```python
movies_data = pd.read_csv('/content/movies.csv')
```

```python
movies_data.head()
```

```python
movies_data.shape
```

```python
selected_features = ['genres', 'keywords', 'tagline', 'cast', 'director']
print(selected_features)
```

```python
for f in selected_features:
  movies_data[f] = movies_data[f].fillna('') ### because to convert to vector, filling null with empty string
```

```python
combined_features = movies_data['genres'] + ' ' + movies_data['keywords'] + ' ' + movies_data['tagline'] + ' ' + movies_data['cast'] + movies_data['director']
```

```python
print(combined_features)
```

```python
"""
TF-IDF : term freq x inverse freq
TF: what is the freq iof that word in that term(sentence)
Inverse document freq: how freq in the whole documents.
we cal: log(total documents/ documents containing word)
rare word = higher IDF

TF-IDF = TF x IDF
meaning:
freq in document(sentence)
rarer across dataset
high importance.
"""
```

```python
vectorizer = TfidfVectorizer()
```

```python
feature_vector = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vector)
```

```python
"""
Cosine Similarity:
it creates to vectors:
1.a = user profile vector
2.b = movie vector
score ranges from 0 to 1
Example:
A = [ 1 1 0]
B = [1 1 0]

now the movie vector that will be created assigned.
cosine similarity = A.B/|A||B|
how one movie is corelated to other.
"""
```

```python
similarity = cosine_similarity(feature_vector)
```

```python
print(similarity)
### 1 because cosine simlarity between the same movies
```

```python
print(similarity.shape)
```

```python
movie_name = str(input("Enter the movie name:"))
```

```python
list_of_all_titles = movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
print('Movies suggested for you : \n')
i = 1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index == index]['title'].values[0]
  if (i < 30):
    print(str(i)+'.', title_from_index)
    i += 1
```
